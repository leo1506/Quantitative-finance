```python
url='https://finance.yahoo.com/news/'
html=req.get(url).content 
soup=bs(html, 'html.parser')
soup = soup.find('div' ,class_ = 'Pos(r) Bgc($bg-content) Bgc($lv2BgColor)! Miw(1007px) Maw(1260px) tablet_Miw(600px)--noRightRail Bxz(bb) Bdstartc(t) Bdstartw(20px) Bdendc(t) Bdends(s) Bdendw(20px) Bdstarts(s) Mx(a)')

```

```python
data = soup.find_all('li')
titles = []
news = []
date_ = []
for i in data:
    date_.append(date.today())
    try:
        news.append(i.text)
        
    except:
        news.append('no disponible')
    try:
        titles.append(i.find('h3').text)
    except:
        titles.append('no disponible')
        
news_data = pd.DataFrame(list(zip(titles,news,date_)),columns = ['titles','news','date'])
news_data = news_data[news_data.titles != 'no disponible']

news_data
```
